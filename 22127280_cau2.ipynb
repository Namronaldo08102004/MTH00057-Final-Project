## **2. Xây dựng Class để giải quyết bài toán**

```python
class BuildingI:
    """
    Since 5 questions are related to each other, so I decided to build a class for initialing a unified set of attributes for answer all questions
    """
    def __init__ (self, numClasses: int, state_2020: list[int], state_2021: list[int]):
        try:
            if (sum(state_2020) == numClasses and sum(state_2021) == numClasses):
                self.numClasses = numClasses
                self.state_2020 = state_2020
                self.state_2021 = state_2021
                self.numStates = len(self.state_2020)
                
                self.P = None
                self.buildTransitionMatrix()
                
                self.r0 = None
                self.defineInitialDistribution()
                
            else:
                raise Exception (f"The total number of rooms in one year is {numClasses}")
        except Exception as e:
            print(f"Error: {e}")
            exit()    
    
    def buildTransitionMatrix (self):
        """
        This function helps to build the transition matrix of the Markov chain problem
        """
        self.P = [[0.0 for _ in range (self.numStates)] for _ in range (self.numStates)]
        state_2020_copy = self.state_2020.copy()
        state_2021_copy = self.state_2021.copy()
        
        for i in range (self.numStates):
            if (i == self.numStates - 1):
                self.P[i][i] = 1
                
            else:
                state_2021_copy[i] = state_2021_copy[i] - ((state_2020_copy[i - 1] - state_2021_copy[i - 1]) if i >= 1 else 0)
                
                self.P[i][i] = state_2021_copy[i] / state_2020_copy[i]
                self.P[i + 1][i] = (state_2020_copy[i] - state_2021_copy[i]) / state_2020_copy[i]
                
    def defineInitialDistribution (self):
        """
        This function helps to define the initial distribution of the problem
        """
        self.r0 = [self.state_2021[i] / self.numClasses for i in range (self.numStates)]
        
    def calculateNumRoomsOfEachState_AtCertainYear (self, year: int) -> list[float]:
        """
        This function helps to calculate the distribution of the problem at a certain year
        """
        def multiplyTwoMatrices (matrix1: list[list[float]], matrix2: list[list[float]]) -> list[list[float]]:
            """
            This function helps to multiply two matrices
            """
            return [[sum([matrix1[i][k] * matrix2[k][j] for k in range (len(matrix1[0]))]) for j in range (len(matrix2[0]))] 
                    for i in range (len(matrix1))] 
            
        def identityMatrix (size: int) -> list[list[float]]:
            """
            This function helps to create an identity matrix with a given size
            """
            return [[int(j == i) for j in range (size)] for i in range (size)]    
            
        def powerOfMatrix (matrix: list[list[float]], exponent: int) -> list[list[float]]:
            """
            This function helps to calculate the power of a square matrix
            """
            if (exponent == 0):
                return identityMatrix(len(matrix))
            
            if (exponent == 1):
                return matrix
            
            temp = powerOfMatrix(matrix, exponent // 2)
            if (exponent & 1):
                return multiplyTwoMatrices(multiplyTwoMatrices(temp, temp), matrix)
            else:
                return multiplyTwoMatrices(temp, temp)
        
        try:
            if (year < 2021):
                raise Exception ("Only calculate distribution at a year after 2021")
            else:
                P_power = powerOfMatrix(self.P, (year - 2021))
                distribution = [sum([P_power[i][j] * self.r0[j] for j in range (self.numStates)]) * self.numClasses for i in range (self.numStates)]
                
                #? Round the distribution
                if (sum([round(distribution[i]) for i in range (len(distribution))]) == self.numClasses):
                    distribution = [round(distribution[i]) for i in range (len(distribution))]
                
                elif (sum([round(distribution[i]) for i in range (len(distribution))]) == self.numClasses - 2): 
                    #? I work with the value 48 because the built-in round() function in Python returns round(12.5) is 12 instead of 13, so to ensure, I will work more with 48
                    distribution = [round(distribution[i]) for i in range (len(distribution))]
                    distribution[len(distribution) - 1] += 1
                    distribution[len(distribution) - 2] += 1
                    
                elif (sum([round(distribution[i]) for i in range (len(distribution))]) == self.numClasses + 2): 
                    distribution = [round(distribution[i]) for i in range (len(distribution))]
                    distribution[0] -= 1
                    distribution[1] -= 1
                    
                elif (sum([round(distribution[i]) for i in range (len(distribution))]) == self.numClasses - 1): 
                    """
                    When the round of the distribution decreases the sum by 1, it means the number of values 
                    which was floored is higher than or equal to the number of values which was ceiled
                    
                    So I will increase the rightmost value which was floored by 1 (The more on the right the value is, 
                    the worse the state is, and this handling will be reasonable because the building will become worse in next years)
                    """
                    maxIndex = -1
                    for i in range (self.numStates):
                        if (round(distribution[i]) - distribution[i] < 0):
                            maxIndex = i
                
                    distribution = [round(distribution[i]) for i in range (len(distribution))]
                    distribution[maxIndex] += 1
                    
                elif (sum([round(distribution[i]) for i in range (len(distribution))]) == self.numClasses + 1):
                    """
                    When the round of the distribution increases the sum by 1, it means the number of values 
                    which was ceiled is higher than or equal to the number of values which was floored
                    
                    So I will decrease the leftmost value which was ceiled by 1 (The more on the left the value is, 
                    the better the state is, and this handling will be reasonable because the building will become worse in next years)
                    """
                    maxIndex = self.numStates
                    for i in range (self.numStates - 1, -1, -1):
                        if (round(distribution[i]) - distribution[i] > 0):
                            maxIndex = i
                
                    distribution = [round(distribution[i]) for i in range (len(distribution))]
                    distribution[maxIndex] -= 1
                    
                return distribution
        except Exception as e:
            print(f"Error: {e}")
            exit()
    
    def predictDowngradation_Within15Years (self) -> list[list[int]]:
        """
        This function helps to record the number of rooms in each state within 15 years starting from the year 2021 (2021-2035)
        """
        statisticalTable = []
        
        for i in range (15):
            distribution = self.calculateNumRoomsOfEachState_AtCertainYear(2021 + i)
            statisticalTable.append(distribution)
            
        return statisticalTable
    
    def determineYearForRebuilding (self) -> int:
        """
        This function helps to determine the year that the building needs to be rebuilt
        """
        year = 2021
        
        while (True):
            distribution = self.calculateNumRoomsOfEachState_AtCertainYear(year)
            if (distribution[len(distribution) - 1] == self.numClasses):
                return year
            else:
                year += 1
```

```python
buildingI = BuildingI(50, [34, 12, 3, 1], [26, 17, 5, 2])
```

## **3. Trả lời câu hỏi**

### $\quad$ **a. Xây ma trận xác suất chuyển**

Gọi X là 1 xích Markov biểu thị trạng thái của CSVC, khi đó, tập trạng thái của X là S = {"tốt", "khá", "trung bình", "xấu"}

Vì trong 1 năm, sự xuống cấp về CSVC (thay đổi trạng thái) của các phòng học diễn ra không qua 1 cấp trong điều kiện PQTTB không thực hiện bất kỳ công tác bào trì, sửa chữa nào, do đó:

$
\begin{cases}
P(X_{n + 1} = "\text{trung bình}" | X_n = "\text{tốt}") = 0 \\
P(X_{n + 1} = "\text{xấu}" | X_n = "\text{tốt}") = 0 \\
P(X_{n + 1} = "\text{tốt}" | X_n = "\text{khá}") = 0 \\
P(X_{n + 1} = "\text{xấu}" | X_n = "\text{khá}") = 0 \\
P(X_{n + 1} = "\text{tốt}" | X_n = "\text{trung bình}") = 0 \\
P(X_{n + 1} = "\text{khá}" | X_n = "\text{trung bình}") = 0 \\
P(X_{n + 1} = "\text{tốt}" | X_n = "\text{xấu}") = 0 \\
P(X_{n + 1} = "\text{khá}" | X_n = "\text{xấu}") = 0 \\
P(X_{n + 1} = "\text{trung bình}" | X_n = "\text{xấu}") = 0 \\
\end{cases}
$

Xét quá trình xuống cấp từ năm 2020 sang năm 2021:
+ Có thể thấy, từ 34 phòng học "tốt" vào năm 2020 xuống còn 26 phòng học "tốt" vào năm 2021, tức đã có 8 phòng học xuống cấp từ tốt thành khá

$\rightarrow 
\begin{cases}
P(X_{n + 1} = "\text{tốt}" | X_n = "\text{tốt}") = \frac{26}{34} \\
P(X_{n + 1} = "\text{khá}" | X_n = "\text{tốt}") = \frac{8}{34}
\end{cases}
$

+ Trong 17 phòng học chất lượng "khá" ở năm 2021, loại bỏ đi 8 phòng học xuống cấp, ta còn 9 phòng học được giữ nguyên chất lượng "khá" so với năm 2020, mà ở năm 2020 có 12 phòng học "khá", nên sẽ có 3 phòng học xuống cấp từ "khá" thành "trung bình"

$\rightarrow 
\begin{cases}
P(X_{n + 1} = "\text{khá}" | X_n = "\text{khá}") = \frac{9}{12} \\
P(X_{n + 1} = "\text{trung bình}" | X_n = "\text{khá}") = \frac{3}{12}
\end{cases}
$

+ Trong 5 phòng học chất lượng "trung bình" ở năm 2021, loại bỏ đi 3 phòng học xuống cấp, ta còn 2 phòng học được giữ nguyên chất lượng "trung bình" so với năm 2020, mà ở năm 2020 có 3 phòng học "trung bình", nên sẽ có 1 phòng học xuống cấp từ "trung bình" thành "xấu"

$\rightarrow 
\begin{cases}
P(X_{n + 1} = "\text{trung bình}" | X_n = "\text{trung bình}") = \frac{2}{3} \\
P(X_{n + 1} = "\text{xấu}" | X_n = "\text{trung bình}") = \frac{1}{3}
\end{cases}
$

Cuối cùng, $P(X_{n + 1} = "\text{xấu}" | X_n = "\text{xấu}") = 1$

Tóm lại:
$$
P = 
\begin{bmatrix}
\frac{26}{34} & 0 & 0 & 0 \\
\frac{8}{34} & \frac{9}{12} & 0 & 0 \\
0 & \frac{3}{12} & \frac{2}{3} & 0 \\
0 & 0 & \frac{1}{3} & 1
\end{bmatrix}
$$
